## Spliting the data 

```python
X=dsn.drop('W',axis=1)
y=dsn['W']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

# Machine Learning Modeling 

## Best score with random state

```python
#Define a function to try different train test folds
def model_select(model):
    max_score=0
    max_state=0
    for x in range(42,105):
        X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=x)
        md=model()
        md.fit(X_train,y_train)
        predict=md.predict(X_test)
        r2score=r2_score(y_test,predict)
        if r2score>max_score:
            max_score=r2score
            max_state=x
    print("Max score is {} with random state {}".format(max_score,max_state))
```

```python
model_select(LinearRegression)
```

```python
model_select(ElasticNet)
```

- Not better than linear 

```python
model_select(Ridge)
```

- better than linear regression 

```python
model_select(Ridge)
```

```python
model_select(KNeighborsRegressor)
```

```python
model_select(RandomForestRegressor)
```

- Better than others but still best is Ridge 

```python
model_select(AdaBoostRegressor)
```

```python
model_select(DecisionTreeRegressor)
```

- 1st = Ridge has best score among all - Max score is 0.9020741688100055 with random state 50
- 2nd = AdaBoostRegressor - Max score is 0.8612902622133126 with random state 70                      
- 3rd = RandomForestRegressor- Max score is 0.8470106682769727 with random state 70      

# Taking our Best model 

## Ridge Regression

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=50)
```

```python
r=Ridge()
r.fit(x_train,y_train)
y_pred=r.predict(x_test)
print('mse ',mean_squared_error(y_test,y_pred))
print('rmse ',mean_squared_error(y_test,y_pred,squared=False))
print('mean absolute error ',mean_absolute_error(y_test,y_pred))
print('r2 score ',r2_score(y_test,y_pred))
```

```python
mse=cross_val_score(r,x,y,scoring='neg_mean_squared_error',cv=7)
print(-mse)
print('mean ',-mse.mean())
```

### lets do some hyperparameter tuning to improve the model accuracy

```python
params_Ridge = {'alpha': [1,0.1,0.01,0.001,0.0001,0] , "fit_intercept": [True, False], "solver": ['svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']}
Ridge_GS = GridSearchCV(r, param_grid=params_Ridge, n_jobs=-1)
Ridge_GS.fit(x_train,y_train)
Ridge_GS.best_params_
```

```python
r1=Ridge(alpha= 1, 
         fit_intercept= 'False',
         solver= 'saga')
r1.fit(x_train,y_train)
y_pred=r1.predict(x_test)
print('mse ',mean_squared_error(y_test,y_pred))
print('rmse ',mean_squared_error(y_test,y_pred,squared=False))
print('mean absolute error ',mean_absolute_error(y_test,y_pred))
print('r2 score ',r2_score(y_test,y_pred))
```
